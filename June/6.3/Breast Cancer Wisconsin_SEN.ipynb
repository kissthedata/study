# 모델 

- 불균형 데이터 처리

```python
from imblearn.over_sampling import SMOTE
x = data.drop(['diagnosis','id'],axis=1)
y = data.loc[:,'diagnosis'].values
sm = SMOTE(k_neighbors = 3) #random_state=0
print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트:', x.shape, y.shape)
x, y = sm.fit_resample(x, y)
print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트:', x.shape, y.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', pd.Series(y).value_counts())
sns.countplot(y)
plt.title('diagnosis', fontsize=10)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y, random_state = 42)
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

하이퍼파라미터 튜닝

   form sklearn.model_selection import GridSearchCV

- 로지스틱 회귀

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param = {'C': randint(low = 0.01, high = 100)}

from sklearn.linear_model import LogisticRegression
#n_iter = 100, 랜덤횟수 디폴트 =10
random_search = RandomizedSearchCV(LogisticRegression(), param_distributions = param, cv = 5, return_train_score = True)
random_search.fit(x_train, y_train)
predicted = random_search.predict(x_test)
score = random_search.score(x_test, y_test)
lr_score = np.mean(score)


print("Best paramneter: {}".format(random_search.best_params_))
print("Best score: {:.3f}".format(random_search.best_score_))
print('Accuracy : %.3f' % (lr_score))
```

- KNN

  k의 수가 1에 가까울 수록 과대 적합
  
  k의 수가 너무 크면 과소 적합을 따름
  
  - 케이스 기반 알고리즘으로 데이터에 따라 결과와 성능이 크게 달라져 머신러닝 분야에는 적합하지 않음 
  - 기저 모델로 다른 알고리즘의 결과와 비교하는 기준으로 사용

```python
param = {'n_neighbors':[1,3,5,7,9]}
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
KNN = GridSearchCV(KNeighborsClassifier(), param, cv = 5)
KNN.fit(x_train, y_train)
predicted = KNN.predict(x_test)
score = KNN.score(x_test, y_test)
kn_score = np.mean(score)

print("Best paramneter: {}".format(KNN.best_params_))
print("Best score: {:.3f}".format(KNN.best_score_))
print('Accuracy : %.3f' % (kn_score))
```

- 나이브베이즈

var_smoothing은 모든 특성치의 최대 분산 비율

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param = {'var_smoothing': randint(low =0, high = 20)}

from sklearn.naive_bayes import GaussianNB
GNB = RandomizedSearchCV(GaussianNB(), param_distributions = param, n_iter =100, cv =5)

GNB.fit(x_train, y_train)
predicted = GNB.predict(x_test)
score = GNB.score(x_test, y_test)
gnb_score = np.mean(score)


print("Best paramneter: {}".format(GNB.best_params_))
print("Best score: {:.3f}".format(GNB.best_score_))
print('Accuracy: %.3f' % (gnb_score))
```

- 의사결정나무

관심대상 집단을 몇 개의 소집단으로 분류, 예측

직관적으로 결과 도식화 가능하지만 안정적이지 못하다는 단점으로 최종 알고리즘 도출 전에 탐색적으로 변수 확인 차원에서 활용

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param = {'max_depth':randint(low=1,high = 20), 'min_samples_leaf':randint(low=1,high= 50)}
from sklearn.tree import DecisionTreeClassifier
tree = RandomizedSearchCV(DecisionTreeClassifier(), param_distributions = param, n_iter =20, cv =20)
tree.fit(x_train, y_train)
predicted = tree.predict(x_test)
score = tree.score(x_test, y_test)
tree_score = np.mean(score)

print('Best Parameter: {}'.format(tree.best_params_))
print('Best score: :{:.3f}'.format(tree.best_score_))
print('Accuracy: %.3f' % (tree_score))
```

- 랜덤 포레스트

```python
from sklearn. model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint
param = {'n_estimators':randint(low=100, high=1000), 'max_features':['auto', 'sqrt', 'log2']}
r_forest = RandomizedSearchCV(RandomForestClassifier(), param_distributions = param, n_iter =20, cv =5)
r_forest.fit(x_train,y_train)
predicted = r_forest.predict(x_test)
score = r_forest.score(x_test, y_test)
rf_score_ = np.mean(score)

print('Best Parameter: {}'.format(r_forest.best_params_))
print('Best score: :{:.3f}'.format(r_forest.best_score_))
print('Accuracy : %.3f' % (rf_score_))
```
