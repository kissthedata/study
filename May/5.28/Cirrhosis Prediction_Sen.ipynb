```python
sns.countplot(data['Stage'])
plt.title('Stage', fontsize=10)
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x = data.drop(['Stage','ID'],axis=1)
y = data.loc[:,'Stage'].values
```

```python
y.shape
```

```python
sm = SMOTE(k_neighbors = 3) #random_state=0
print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트:', x.shape, y.shape)
x, y = sm.fit_resample(x, y)
print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트:', x.shape, y.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', pd.Series(y).value_counts())
```

```python
sns.countplot(y)
plt.title('Stage', fontsize=10)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
r_forest = RandomForestClassifier(criterion = 'entropy', max_depth = 20, n_estimators = 10000)
r_forest.fit(x_train,y_train)
predicted = r_forest.predict(x_test)
score = r_forest.score(x_test, y_test)
rf_score_ = np.mean(score)

print('Accuracy : %.3f' % (rf_score_))
```
